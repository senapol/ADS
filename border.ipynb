```python
# imports
import plotly.express as px
import geopandas as gpd
import shapely.geometry
import numpy as np
import json
```

```python
#lon, lat
with open('data/border.json', 'r') as file:
    data = json.load(file)

coords = data['features'][0]['geometry']['coordinates'][0][0]

lats = []
lons = []

for coord in coords:
    lat = coord[1]
    lon = coord[0]
    lats.append(lat)
    lons.append(lon)

print('Plotting map with Ukraine borders')
fig = px.line_geo(lat=lats, lon=lons)

fig.show()
```

```python
# Load neural network frontline data
import json
with open('data/nn_frontline_data.json', 'r') as file:
    nn_data = json.load(file)

# Specify the date to extract points
date = "2023-04-17"  # Replace with the desired date

# Extract and flatten the points for the specified date
frontline_points = []
if date in nn_data:
    for segment in nn_data[date]:
        for coord in segment:
            if len(coord) == 2:  # Ensure valid [lon, lat] format
                frontline_points.append(coord)
            elif len(coord) == 1:  # Handle incomplete coordinates
                frontline_points.append([coord[0], 0])  # Placeholder latitude

# Separate lons and lats for plotting
lons = [point[0] for point in frontline_points]
lats = [point[1] for point in frontline_points]

# Plot the points on a map
print(f"Plotting frontline points for {date}")
fig = px.scatter_geo(lat=lats, lon=lons, title=f"Frontline Points for {date}")
fig.update_geos(fitbounds="locations")  # Automatically zoom to points
fig.show()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoapi", timeout=10)
sample = (str(lats[0]), str(lons[0]))
print(sample)
location = geolocator.reverse(f"{str(lats[0])}, {str(lons[0])}", language='en')  # Example coordinates
print(location.address)
if 'crimea' in location.address.lower():
    print('In Crimea')
```

```python
coords_without_occupied = []
removed = 0
for i in range(len(coords)):
    lat = coords[i][1]
    lon = coords[i][0]
    sample = (str(lat), str(lon))
    location = geolocator.reverse(f"{str(lat)}, {str(lon)}", language='en')
    if all(region not in location.address.lower() for region in ['crimea', 'donetsk', 'luhansk', 'russia']):
        coords_without_occupied.append(coords[i])
    else:
        print(f'Removed point in region {location.address}')
        removed += 1

print(f'Removed {removed} points occupied by Russia')
```
